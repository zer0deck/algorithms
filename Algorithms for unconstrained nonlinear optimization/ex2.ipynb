```python
#libs initialization
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import random
import tkinter as tk
import tkinter.ttk as ttk
from tkinter import *
import scipy
from scipy import optimize
from scipy.optimize import minimize
```

```python
#parameters
alpha = random.uniform (0, 1)
beta = random.uniform (0, 1)
arr_ex_1_2_x = []
arr_ex_3_x = []
arr_ex_1 = []
arr_ex_2 = []
arr_ex_3 = []
arr_1_ex_4 = []
arr_2_ex_4 = []
y_ex_4 = []
x_ex_4 = []
b = []
a_1=0
b_1=0
a_2=0
b_2=0
a_3 = 0.05
b_3 = 0.05
b_3_temp = 10000
a_4 = 0.05
b_4 = 0.05
b_4_temp = 10000
```

```python
#functions creation
def f_ex_1(x):
    y = x ** 3
    return (y)
def f_ex_2(x):
    y = np.abs (x - 0.2)
    return (y)
def f_ex_3(x):
    y = x * np.sin (1 / x)
    return (y)
def f_ex_4 (x):
    y = alpha * x + beta + random.normalvariate (0.5, 0.125)
    return y
```

```python
#Arrays filling loops
for i in range (0, 1001, 1):
    arr_ex_1.append( f_ex_1 (i / 1000))
    arr_ex_1_2_x.append (i/1000)
    arr_ex_2.append( f_ex_2 (i / 1000))
    arr_ex_3_x.append (0.99 / 1000 * i + 0.01)
    arr_ex_3.append( f_ex_3 (0.99 / 1000 * i + 0.01))
for k in range (0, 101, 1):
    y_ex_4.append( f_ex_4 (k / 100 ))
    x_ex_4.append(k/100)
```

```python
#algorithms
def bruteforce_alg(function, a, b):
    n = int((b - a) / 0.001)
    rounds = 0
    evaluation = 0
    ymin = float("inf")
    for k in range(n+1):
        rounds += 1
        x = a + k * (b - a) / n
        evaluation += 1
        y = function( x )
        if y < ymin:
            ymin = y
            xmin = x     
    return ymin, rounds, evaluation, xmin

def dich_alg(function, a, b):
    beta = 0.0009
    rounds = 0
    evaluation = 0
    while np.abs(a-b) > 0.001:
        xmin = (a + b - beta)/2
        xmax = (a + b + beta)/2
        rounds+=1
        evaluation+=2
        if function(xmin) <= function(xmax):
            b = xmax
        else:
            a = xmin
    n=(a+b)/2
    return function((a+b)/2), rounds, evaluation, n

def golden_alg(function, a, b):
    x1 = a + ( 3 - np.sqrt(5) ) / 2 * ( b - a )
    x2 = b + ( np.sqrt(5) - 3 ) / 2 * ( b - a )
    rounds = 0
    evaluation = 1
    while np.abs(a-b) > 0.001:
        rounds+=1
        evaluation+=1
        if function(x1) <= function(x2):
            b = x2
            x2 = x1
            x1 = a + ( 3 - np.sqrt(5) ) / 2 * ( b - a )
            n=b
        else:
            a = x1
            x1 = x2
            x2 = b + ( np.sqrt(5) - 3 ) / 2 * ( b - a )
            n=a
    return function((a+b)/2), rounds, evaluation, n
```

```python
#brutforce algorithm for 2 part
for a in range(0, 1000):
    a = a / 1000
    for b in range(0, 1000):
        b1 = b / 1000
        b2 = - b / 1000
        s1 = 0
        s2 = 0
        for k in range(0, 101):
            s1 += (a * x_ex_4[k] + b1 - y_ex_4[k]) ** 2
            s2 += (a / (1 + b2 * x_ex_4[k]) - y_ex_4[k]) ** 2
        arr_1_ex_4.append([s1, a, b1])
        arr_2_ex_4.append([s2, a, b2])
minimum_1=min(x1[0] for x1 in arr_1_ex_4)
minimum_2=min(x2[0] for x2 in arr_2_ex_4)
for sublist in arr_1_ex_4:
    if sublist[0] == minimum_1:
        a_1 = sublist[1]
        b_1 = sublist[2]
        break
for sublist in arr_2_ex_4:
    if sublist[0] == minimum_2:
        a_2 = sublist[1]
        b_2 = sublist[2]
        break
```

```python
#gauss alghorithm for 2 part
while 0 == 0:
    ex_4_gauss_arr_1 = []
    ex_4_gauss_arr_2 = []
    for a in range(0, 1000):
        a = a / 1000
        s = 0
        for k in range(0, 101):
            s += (a * x_ex_4[k] + b_3 - y_ex_4[k])**2
        ex_4_gauss_arr_1.append(s)
    a_3_temp = ex_4_gauss_arr_1.index(min(ex_4_gauss_arr_1)) / 1000
    if abs(a_3 - a_3_temp) < 0.001 and abs(b_3 - b_3_temp) < 0.001:
        break
    a_3 = a_3_temp
    for b in range(0, 1000):
        b = b / 1000
        s = 0
        for k in range(0, 101):
            s += (a_3 * x_ex_4[k] + b - y_ex_4[k])**2
        ex_4_gauss_arr_2.append(s)
    b_3_temp = ex_4_gauss_arr_2.index(min(ex_4_gauss_arr_2)) / 1000
    if abs(b_3 - b_3_temp) < 0.001 and abs(a_3 - a_3_temp) < 0.001:
        break
    b_3 = b_3_temp
while 0 == 0:
    ex_4_gauss_arr_1 = []
    ex_4_gauss_arr_2 = []
    for a in range(0, 1000):
        a = a / 1000
        s = 0
        for k in range(0, 101):
            s += (a / (1 + b_4 * x_ex_4[k]) - y_ex_4[k])**2
        ex_4_gauss_arr_1.append(s)
    a_4_temp = ex_4_gauss_arr_1.index(min(ex_4_gauss_arr_1)) / 1000
    if abs(a_4 - a_4_temp) < 0.001 and abs(b_4 - b_4_temp) < 0.001:
        break
    a_4 = a_4_temp
    for b in range(0, 1000):
        b = - b / 1000
        s = 0
        for k in range(0, 101):
            s += (a_4 / (1 + b * x_ex_4[k]) - y_ex_4[k])**2
        ex_4_gauss_arr_2.append(s)
    b_4_temp = ex_4_gauss_arr_2.index(min(ex_4_gauss_arr_2)) / -1000
    if abs(b_4 - b_4_temp) < 0.001 and abs(a_4 - a_4_temp) < 0.001:
        break
    b_4 = b_4_temp
```

```python
def fun1(ab):
    a, b = ab
    s = 0
    for k in range(0, 101):
        s += (a * x_ex_4[k] + b - y_ex_4[k]) ** 2
    return s
def fun2(ab):
    a, b = ab
    s = 0
    for k in range(0, 101):
        s += (a / (1+x_ex_4[k]*b)- y_ex_4[k]) ** 2
    return s
result1 = minimize(fun1,[0.3, 0.3], method='nelder-mead', options={'xatol': 0.001,'disp': True})
result2 = minimize(fun2,[0.3, 0.3], method='nelder-mead', options={'xatol': 0.001,'disp': True})
```
